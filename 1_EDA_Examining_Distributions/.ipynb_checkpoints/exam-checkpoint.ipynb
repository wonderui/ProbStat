```python
import pandas as pd

data = pd.read_excel('drinking.xls', names = ['Gender', 'Alcohol', 'Height', 'Cheat'])

data
```

```python
alcohol_not_star = data['Alcohol'].index[data['Alcohol'] != '*']
alcohol_not_star
```

```python
alcohol = data.iloc[alcohol_not_star]
alcohol
```

```python
alcohol['Alcohol']
```

```python
alcohol['Alcohol'].mean()
```

```python
alcohol['Alcohol'].std()
```

```python
alcohol['Alcohol'].quantile(q = 0.25), alcohol['Alcohol'].quantile(q = 0.5), alcohol['Alcohol'].quantile(q = 0.75)
```
